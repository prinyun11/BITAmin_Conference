```python
def collector(driver, data, k):
    driver = webdriver.Chrome("driver/chromedriver 3")
    # k번째 회차의 베스트도전 웹툰 웹사이트 오픈
    url = 'https://comic.naver.com/bestChallenge/list.nhn?titleId=722761&page='+str(k)
    driver.get(url)
    
    time.sleep(1)
    
    # 댓글창에 해당하는 프레임 선택
    driver.switch_to.frame('commentIframe')
    
    WebDriverWait(driver, 10)
    
    # 회차,이름,날짜,아이디,댓글,좋아요/싫어요 수 전체 수집
    num = k
    names = driver.find_elements_by_css_selector('.u_cbox_nick')
    nicks = driver.find_elements_by_css_selector('.u_cbox_id')
    comments = driver.find_elements_by_css_selector('.u_cbox_text_wrap')
    
    # 총 n개의 댓글이 수집되므로 해당 댓글 수만큼 for문을 돕니다.
    for i in range(len(names)):
        # for문을 돌며, 댓글별 회차,이름,아이디,댓글 저장
        temp = []
        temp.append(num)
        temp.append(names[i].text)
        temp.append(nicks[i].text)
        temp.append(comments[i].text)

        temp = pd.DataFrame(data=[temp], columns=data.columns)
        data = pd.concat([data,temp])

    # 열어두었던 driver를 닫습니다.
    driver.close()

    print(str(num) + '화 베댓 수집 완료')

    return data
```

```python
for k in range(36):
    data = collector(driver, data, k)
    
print(data)
```
